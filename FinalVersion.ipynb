# ANALYZING COVID-19 DATA

```python
%matplotlib inline
import pandas
```

# importing data

```python
df = pandas.read_excel('s3://brotemarkle-ia241/COVID-19-geographic-disbtribution-worldwide-2020-05-03.xlsx')
df
```

# average number of deaths per day

```python
avg_deaths_per_day = df.groupby('day').mean()['deaths']
avg_deaths_per_day
```

```python
avg_deaths_per_day.plot()
```

# daily average deaths per country

```python
avg_deaths_per_countriesAndTerritories = df.groupby('countriesAndTerritories').mean()['deaths']
avg_deaths_per_countriesAndTerritories
```

```python
avg_deaths_per_countriesAndTerritories.nlargest(10).plot.bar()
```

# daily average of deaths per continent

```python
avg_deaths_per_continentExp = df.groupby('continentExp').mean()['deaths']
avg_deaths_per_continentExp 
```

```python
avg_deaths_per_continentExp.plot.bar()
```
