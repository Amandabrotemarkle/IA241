```python
import pandas
```

```python
df = pandas.read_excel('s3://brotemarkle-ia241/house_price.xls')
df
```

```python
df['unit_price']=df['price']/df['area']
```

```python
df
```

```python
df['years']=pandas.to_datetime(df['built_in'],format='%Y')
```

```python
df
```

```python
avg_price_per_year = df.groupby('years').mean()['price']
avg_price_per_year
```

```python
avg_price_per_year.plot()
```

```python
avg_unit_price_per_year = df.groupby('years').mean()['unit_price']

avg_unit_price_per_year
```

```python
avg_unit_price_per_year.plot()
```
